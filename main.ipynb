**Özellik Değişkenleri Arasındaki İlişkinin Korelasyon Matrisi İle İncelenmesi**  
Korelasyon matrisi, farklı özellikler arasındaki ilişkiyi ölçer ve bu ilişkinin gücünü belirtir. Pozitif bir korelasyon, bir özelliğin artışıyla diğer özelliğin de arttığını gösterirken, negatif bir korelasyon, bir özelliğin artışıyla diğerinin azaldığını gösterir. Korelasyon katsayısı -1 ile +1 arasında değişir, -1'e yaklaştıkça negatif korelasyon gücü artar, +1'e yaklaştıkça pozitif korelasyon gücü artar, 0 ise iki özellik arasında bir ilişkinin olmadığını belirtir.

Verilen korelasyon matrisine dayanarak, kullanabileceğiniz en iyi özellikler şunlar olabilir:

- thumbs_up (beğeniler): Tarifleri beğenen kullanıcılar genellikle tarifi beğenmişlerdir. Bu, popüler tarifleri belirlemede ve kullanıcıların tercihlerini anlamada yardımcı olabilir.

- best_score (en iyi skor): En iyi skor, tarifin genel memnuniyet seviyesini gösterir. Yüksek bir en iyi skor, tarifin kaliteli olduğunu ve kullanıcıların genel olarak memnun kaldığını gösterebilir.

- reply_count (yanıt sayısı): Tarife verilen yanıt sayısı, kullanıcıların tartışmaya dahil olma ve etkileşimde bulunma istekliliğini yansıtabilir. Daha fazla yanıt genellikle daha aktif bir topluluğu işaret eder.

- user_reputation (kullanıcı itibarı): Kullanıcı itibarı, bir kullanıcının platformda ne kadar aktif ve katkıda bulunduğunu gösterir. Yüksek itibara sahip kullanıcılar genellikle diğer kullanıcılar tarafından değerli bulunur ve etkileşimi teşvik edebilir.

```python

numerical_vars = ['user_reputation', 'reply_count', 'thumbs_up', 'thumbs_down', 'stars', 'best_score']
correlation_matrix = df[numerical_vars].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Korelasyon Matrisi')
plt.show()
```

**Özelliklerin İki Boyutlu Düzlemde Dağılımının Görselleştirilmesi (Pairplot)**

seaborn kütüphanesinin pairplot fonksiyonunu kullanarak veri çerçevesindeki sayısal değişkenler arasındaki ilişkiyi görselleştirir. Her bir sayısal değişken çifti için bir scatter plot çizilir ve diyagonalde ise her bir değişkenin dağılımını gösteren bir histogram yer alır.

- thumbs_up ve best_score: Beğeniler ile en iyi skor arasında pozitif bir ilişki gözlemleniyor gibi görünüyor. Yani, bir tarifin daha fazla beğeni alması genellikle daha yüksek bir en iyi skora sahip olmasıyla ilişkilidir. Bu, kullanıcıların genel olarak memnun oldukları tarifleri beğendiklerini ve yüksek bir skor verme eğiliminde olduklarını gösterebilir.

- reply_count ve user_reputation: Yanıt sayısı ile kullanıcı itibarı arasında bir ilişki olduğu görülüyor gibi. Yüksek bir kullanıcı itibarına sahip olan kullanıcılar genellikle daha fazla yanıt alıyor gibi görünmektedir. Bu, platformdaki aktif ve katkıda bulunan kullanıcıların topluluk içinde daha fazla etkileşim aldığını gösterebilir.

- stars ve best_score: Yıldız derecelendirmesi ile en iyi skor arasında da bir ilişki görünüyor. Bu da, yüksek bir yıldız derecelendirmesine sahip olan tariflerin genellikle yüksek bir en iyi skora da sahip olduğunu gösterir. Kullanıcıların genel memnuniyetini yansıtan bu iki özellik arasındaki ilişki doğal olarak beklenebilir.


```python

sns.pairplot(df[numerical_vars])
plt.show()
```

**Tüm Zamanlar Star-Count Dağılım Grafiği**
Grafik incelendiğinde, tariflere verilen puanların yüksek olduğu gözlemlenmiştir ve yoğun olarak 5 yıldız verilmiştir.

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['stars'], bins=5, kde=True)
plt.title('Yıldız-Sayı Dağılım Grafiği')
plt.show()
```

**Aylara Göre İndekslenmiş Yıldız Dağılım Grafiği**
Grafik incelendiğinde, 2022 yılının 6. ayında tariflerin en çok etkileşim aldığı ve puanlandığı gözlemlenmiştir. Puan dağılımı önceki grafikte ifade edildiği gibi yoğun olarak 5 yıldız olarak değerlendirilmiştir.

```python
# Zaman serisi analizi - stars değişkeninin sayısı
monthly_stars_count = df.resample('M', on='created_at')['stars'].count()

plt.figure(figsize=(10, 6))
monthly_stars_count.plot(color='blue')
plt.xlabel('Zaman')
plt.ylabel('Puan Sayısı')
plt.title('Aylık Puan Sayısı')
plt.grid(True)
plt.show()
```

**Kullanıcı İtibarına Göre Tüm Tariflerin Beğenilme Dağılım Grafiği**
Kullanıcı itibar puanlarına göre yapılan yorumlara pozitif etkileşim (thumb up) değerlendirmeleri düşük itibar puanlı kullanıcılar tarafından domine edilmiş ve sağlandığı gözlenmiş ve dağılım grafikte belirtilmiştir.

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='user_reputation', y='thumbs_up', data=df)
plt.title('User Reputation vs Thumbs Up')
plt.show()
```

**Yapılan Yorumların Kelime Sayısına Göre Best Score Dağılım Grafiği**
Yapılan yorumlardaki kelime sayısına oranla en iyi skor arasında güçlü bir ilişki görülmüş ve grafikte bu ilişki sayısal olarak ifade edilmiştir.

```python
# Create the 'comment_length' column
df['comment_length'] = df['text'].str.len()
plt.figure(figsize=(12, 6))

# Yorum uzunlukları dağılımı
sns.histplot(df['comment_length'], bins=30, kde=True, color='skyblue', label='Yorum Uzunluğu')

# Best score değerlerinin dağılımı
sns.histplot(df['best_score'], bins=5, kde=True, color='orange', label='Best Score')

plt.title('Yorum Uzunluğu ve Best Score Dağılımı')
plt.xlabel('Değer')
plt.ylabel('Frekans')
plt.legend()
plt.show()
```

**En Çok Beğenilen İlk 10 Tarif ve Puanları**
En çok beğenilen Tarif cheese burger soup olarak belirlenmiş ve en çok beğenilen ilk 10 yemeğin ortalama yıldız puanı ile birlikte grafikte belirtilmiştir.

```python
top_recipes_freq = df['recipe_name'].value_counts().nlargest(10)

star_counts = df.groupby('recipe_name')['stars'].mean().loc[top_recipes_freq.index]

plt.figure(figsize=(10, 6))
ax = top_recipes_freq.plot(kind='bar', color='skyblue', label='Yorum Sayısı')

for i, count in enumerate(star_counts):
    ax.text(i, top_recipes_freq.iloc[i], f'{count:.2f}', ha='center', va='bottom')

plt.xlabel('Tarif Adı')
plt.ylabel('Ortalam Puanı')
plt.title('En Çok Beğenilen 10 Tarif ve Ortalama Yıldız Sayısı')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.show()
```

**Tariflere En Sık Yorum Yapan İlk 10 Aktif Kullanıcı**
Tariflere en sık yorum yapan itibar puanı yüksek kullanıcılar, yorum yapma sıklığına göre sıralanmış ve en aktif 10 kullanıcı belirlenmiştir.

```python

top_users_freq = df['user_name'].value_counts().nlargest(10)
plt.figure(figsize=(10, 6))
top_users_freq.plot(kind='bar', color='salmon')
plt.xlabel('Kullanıcı Adı')
plt.ylabel('Yorum Sayısı')
plt.title('En Aktif 10 Kullanıcı')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**En Çok Etkileşim Alan İlk 5 Kullanıcı Yorumu, Kullanıcı Bilgisi ve Tarif İsmi**
Genel olarak pozitif veya tarifin içeriği hakkında detaylar bulunduran ve kullanıcı memnuniyetini belirten yorumların en fazla etkileşimi aldığı gözlemlenmiş ve çıktıda görüldüğü gibi ifade edilmiştir.

```python
top_5_comments = df.nlargest(5, 'thumbs_up')
top_5_comments_with_info = top_5_comments[['recipe_name', 'user_name', 'text']]
print(top_5_comments_with_info)
```

**Zamana Bağlı Olarak Yapılan Yorum Sayısı ve Tarife Verilen Ortalama Yıldız Değerleri**
Tüm kullanıcıların zaman bağlı olarak aylık ölçekteki zaman serisi grafiğinde, tariflere yapılan yorum sayıları ve tariflere verilen puanlar grafikte belirtilmiş ve zamana bağlı tariflere olan etkileşim grafikte ifade edilmiştir.


```python
average_stars_over_time = df.groupby(pd.Grouper(key='created_at', freq='M'))['stars'].mean()
comment_count_resample = df.groupby(pd.Grouper(key='created_at', freq='M'))['text'].count()

plt.figure(figsize=(10, 6))
average_stars_over_time.plot(color='green', label='Ortalama Puan')
comment_count_resample.plot(color='purple', label='Yorum Sayısı', secondary_y=True)
plt.xlabel('Zaman')
plt.title('Zaman İçinde Ortalama Puanlar ve Yorum Sayıları')
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
```
